```python
# データの読み込み
df_org = pd.read_csv('./data/ks-projects-201801.csv')
```

```python
# # 成功（successful）と失敗（failed）以外のデータを除外する
# df = df_org[(df_org['state'] == 'successful') | (df_org['state'] == 'failed')]
```

```python
df = df_org
```

```python
list(df.columns)
```

```python
list(df.state.unique())
```

```python
df = df.query('state != "live"')
df = df.assign(isSuccessful = (df['state']=='successful').astype(int))
nop = len(df)
nos = len(df.query('isSuccessful == 1'))
ros = nos/nop
print("Number of projects: ", nop, " number of successful projects: " ,nos, " ratio of success: ", ros)
```

```python
df['deadline'] = pd.to_datetime(df['deadline'])
df['launched'] = pd.to_datetime(df['launched'])
```

```python
df = df.assign(lhour=df.launched.dt.hour,
               lday=df.launched.dt.day,
               lmonth=df.launched.dt.month,
               lyear=df.launched.dt.year)
```

```python
df = df[(df['launched'] >= '2009-04-28')]
```

```python
df = df.assign(dhour=df.deadline.dt.hour,
               dday=df.deadline.dt.day,
               dmonth=df.deadline.dt.month,
               dyear=df.deadline.dt.year)
```

```python
df = df.assign(total_year= df.dyear-df.lyear)
df = df.assign(total_month = df.total_year*12 + (df.dmonth-df.lmonth))
df = df.assign(total_day = df.total_month * 30 + (df.dday - df.lday))
df = df.assign(total_hour = df.total_day * 24 + (df.dhour - df.lhour))
```

```python
df = df.query('total_hour >= 0')
df.head()
```

```python
low = df['goal'].quantile(0.3)
mid = df['goal'].quantile(0.6)
high = df['goal'].quantile(0.9)

conditions = [(df['goal']<low),((df['goal']>=low) & (df['goal']<mid)),((df['goal']>=mid) & (df['goal']<high)),(df['goal']>=high)]
tags = ['t0_low', 't1_medium','t2_high','t3_extreme']

df = df.assign(categorical_goal = np.select(conditions,tags))
```

```python
low = df['backers'].quantile(0.3)
mid = df['backers'].quantile(0.6)
high = df['backers'].quantile(0.9)


conditions = [(df['backers']<low),((df['backers']>=low) & (df['backers']<mid)),((df['backers']>=mid) & (df['backers']<high)),(df['backers']>=high)]
tags = ['t0_low', 't1_medium','t2_high','t3_extreme']

df = df.assign(categorical_backers = np.select(conditions,tags))

df.head()
```

```python
groupableData = df.drop(columns = ['ID', 'name','goal','backers', 'deadline','dyear','lyear','dmonth','lmonth','dday','lday','dhour','lhour','launched','usd_goal_real','pledged','state','usd_pledged_real','usd pledged'])
groupableData.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df = groupableData.apply(encoder.fit_transform)
df.head()
```

```python
df_ks = df
df_ks.head()
```

```python
df_ks = df_ks.drop(columns = ['categorical_backers', 'currency'])
```

```python
frac = 0.1
size = int(len(df_ks)*frac)

trainSet = df_ks[:-2*size]
validationSet = df_ks[-2*size: -size]
testSet = df_ks[-size:]
```

```python
import lightgbm as lgb
features = trainSet.columns.drop('isSuccessful')

dtrain = lgb.Dataset(trainSet[features], label=trainSet['isSuccessful'])
dvalid = lgb.Dataset(validationSet[features], label=validationSet['isSuccessful'])

param = {'num_leaves': 64, 'objective': 'binary'}
# param['metric'] = 'auc'
param['metric'] = 'binary_logloss'
num_round = 1000
bst = lgb.train(param, dtrain, num_round, valid_sets=[dvalid], early_stopping_rounds=10, verbose_eval=False)
```

```python
from sklearn import metrics
ypred = bst.predict(testSet[features])
# score = metrics.roc_auc_score(testSet['isSuccessful'], ypred)
score = metrics.accuracy_score(testSet['isSuccessful'], ypred.round(0))

print(f"Test accuracy_score: {score}")
```
