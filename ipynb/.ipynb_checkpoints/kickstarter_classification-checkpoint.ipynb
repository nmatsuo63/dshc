```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# 以下に必要なライブラリのインポート文を記入

```

## 前処理

```python
# データの読み込み
df = pd.read_csv('ks-projects-201801.csv')

# 成功（successful）と失敗（failed）以外のデータを除外する
df = df[(df['state'] == 'successful') | (df['state'] == 'failed')]
```

## モデルの学習と評価

```python
# 訓練データとテストデータを8:2に分割
# ※評価を揃えるため df 以外の引数は変更しないこと
train, test = train_test_split(df, test_size=0.2, random_state=1234)
```

```python
# 以下に学習用のコードを記入


```

```python
# 訓練データに対する予測
y = np.ones(len(train['state'])) # ダミーの訓練データの目的変数, 適宜変更せよ
y_pred = np.ones(len(train['state'])) # ダミーの予測結果, 自由に変更してよい

# テストデータに対する正答率（accuracy）を計算
# ※ 評価を揃えるため, accuracy_scoreは変更しないこと
acc_train = accuracy_score(y_true=y, y_pred=y_pred)

# 正答率を小数点第二位まで表示した結果
print(f"訓練データの正答率: {100 * acc_train:.2f} %")
```

```python
# テストデータに対する予測
y = np.ones(len(test['state'])) # ダミーのテストデータの目的変数, 適宜変更せよ
y_pred = np.ones(len(test['state'])) # ダミーの予測結果, 自由に変更してよい

# テストデータに対する正答率（accuracy）を計算
# ※ 評価を揃えるため, accuracy_scoreは変更しないこと
acc_test = accuracy_score(y_true=y, y_pred=y_pred)

# 正答率を小数点第二位まで表示した結果
# この結果を報告してください
print(f"テストデータの正答率: {100 * acc_test:.2f} %")
```
